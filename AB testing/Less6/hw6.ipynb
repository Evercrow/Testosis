### Задача 1
*Продакт на главной mail.ru решил протестировать в рекомендательной ленте контента вместо карточек со статьями видеоплеер с короткими видео. Нынешний таймспент на юзера в день в среднем равен 25 минут, а стандартная ошибка (SD) равна 156. Мы предполагаем, что в новой версии таймспент на юзера в день изменится на 10%. Средний трафик 20000 человек в день. Посчитайте сколько дней необходимо держать эксперимент при alpha = 5% и beta = 20% .*


**Решение:**  
Задача на определение размера выборки по заданной дисперсии(отклонению) для непрерывных величин(время).  
Дано:  
Метрика timespent, ожидаемое среднее время контрольной группы = 25 мин,  
Standart Deviation = 156,  
Uplift (relative ) = 10%,  ожидаемое среднее время контрольной группы = 25+ 0.1*25 = 27.5
$\alpha$ и $\beta$ стандартные 5 и 20%  
Найдем подходящий калькулятор и подсчитаем необходимое для теста количество юзеров:  
!["Расчет на https://statulator.com/SampleSize/ss2M.html"](ab_hw6-1.jpg)

Получаем общее число юзеров в 122248 , которые поделятся на две группы. Поделив его на среднее число юзеров в день , получим нужную длительность для эксперимента в днях для наших показателей трафика: 
test_duration = 122248/20000 = **6,1124** , округляем до большего целого, чтобы не искажать  статистику разницей между временем суток(вечер-утро и т.п.) и не терять в точности, получаем ровно **неделю**


### Задача 2
*Наша продуктовая команда в ecommerce магазине планирует запустить тест, направленный на ускорение загрузки сайта. Одна из основных метрик bounce rate в GA = 40%. Мы предполагаем, что при оптимизации сайта она изменится минимум на 20%.Средний трафик 4000 человек в день. Посчитайте сколько нам нужно дней держать эксперимент при alpha = 5% и beta = 20%*

**Решение:**

Задача схожая с предыдущей, но метрика у нас теперь не непрерывная(количественная), а больше качественная (пользователь зашел  по адресу или нет) величина. Используем калькулятор для аналогичной задачи по расчету изменения конверсии(пропорциональной величины). В условии неясно, имеется ли в виду изменение величины в абсолютном формате или в относительном, поэтому посчитаем для обоих вариантов:  
!["https://www.evanmiller.org/ab-testing/sample-size.html"](ab_hw6-2.jpg)

1) test_duration = 592*2/4000 = 0.296 дня = 7.104 часов
2) test_duration = 94*2/4000 = 0.047 дня = 1.128 часов 
В обоих случаях стоит брать **один полный день** на тест, если у нас нет никаких возможностей сделать корректировки по разночасовым факторам трафика (метрики среднего числа посетителей между 10 и 18 часами, например)
