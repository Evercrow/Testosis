### Задание 2
*Сравниваем метрику конверсия в покупку. Размер выборки - 10000 элементов в каждой группе . Какой статистический критерий тут лучше всего подойдёт и почему ?*


По определению, конверсия - это процентное соотношение числа пользователей, совершивших целевое действие, к общему числу пользователей, посетивших сайт. Следовательно, это качественная величина, по 2 независимым выборкам с условно нормальным распределением(число пользователей 10000 или больше) **подойдет криттерий Хи-квадрата Пирсона**.
Результаты теста так же можно оценить по **z - критерию для оценки разности долей**  формула:
$$z = \frac{\bar{p_1}-\bar{p_2}}{\sqrt{p(1-p)(\frac{1}{n_1}+\frac{1}{n_2})}} \text{ ,где  }p = \frac{m_1+m_2}{n_1+n_2}$$
отношение числа благоприятных исходов m1, m2 к размеру выборок n1,n2  

Бернулли не подойдет, так как размеры выборок большие(>5)
